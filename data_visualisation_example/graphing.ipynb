```python
# Import necessary packages for this script
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Importing an example CSV file
sales_df = pd.read_csv("company_sales_data.csv")
```

Example 1: Total profit of all months and show it using a line plot

```python
plt.figure()
sns.lineplot(x="month_number", y="total_profit", data=sales_df, label="Sales")
plt.title("Total Profit for all Months")
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
plt.xticks(sales_df["month_number"])
plt.legend()
plt.show()
```

Example 2: Total profit of all months and show line plot

```python
plt.figure()
sns.lineplot(x="month_number", y="total_profit", data=sales_df, color="red", linewidth=3, label="Profit Data of Last Year", linestyle="--", marker="o", markerfacecolor='k')
plt.title("Total Profit for all Months")
plt.xlabel("Month Number")
plt.ylabel("Sold Units Number")
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.xticks(sales_df["month_number"])
plt.legend(loc="lower right")
plt.show()
```

Example 3: Read all product sales data and show it  using a multiline plot

```python
plt.figure()
sns.lineplot(data=sales_df[["facecream", "facewash", "toothpaste", "bathingsoap", "shampoo", "moisturizer"]])
plt.xlabel("Month Number")
plt.ylabel("Sales Units in Number")
plt.title("Sales Data")
plt.show()
```

Example 4: Read toothpaste sales data of each month and show it using a scatter plot

```python
plt.figure()
toothpaste = sales_df["toothpaste"]
sns.scatterplot(data=toothpaste, label="Toothpaste Sales Data")
plt.xlabel("Month Number")
plt.ylabel("Number of Units Sold")
plt.title("Toothpaste Sales Data")
plt.legend()
plt.xticks(sales_df["month_number"])
plt.grid(True, linestyle="--")
plt.show()
```

Example 5: Bathing soap sales data for all months

```python
plt.figure()
sns.barplot(data=sales_df, x="month_number", y="bathingsoap")
plt.xlabel("Month Number")
plt.ylabel("Sales units")
plt.title("Bathing Soap Sales Data")
plt.grid(True, linestyle="--", linewidth=1)
plt.show()
```

Example 6: Read the total profit of each month and show it using the histogram to see the most common profit ranges

```python
plt.figure()
sns.histplot(data=sales_df["total_profit"], bins=6, label="Profit Data", binrange=[150000, 450000])
plt.ylabel("Actual Profit in USD")
plt.xlabel("Profit Range in USD")
plt.title("Profit Data")
plt.legend()
plt.show()
```
